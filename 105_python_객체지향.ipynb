## 상속(inheritance)
 - class에서 이름 뒤에 괄호는 상속이다.
 - python의 모든 객체는 object를 상속받는다
 - python3 부터는 (object)는 생략가능하다. 즉 지금까지 class를 정의할때 object내부적으로 상속하고 있었다.

`````python
# 상속을 해주는 클래스 : 부모,수퍼,기본 클래스
# 상속을 받는 클래스   : 자식,서브,유도 클래스

# 부모클래스
class Parent:
  x = 'parent x' #클래스 변수
  y = 'parent y'

  @classmethod          # 상속받는 클래스에 부모 또는 자신의 자원을 호출해서 사용할수 있도록 제공해주는 기능
  def output(cls):      # 각 클래스에 있는 값을 리턴             # 호출할떄 객체값 안넘김(마지막줄)
    return cls.x

  def test(self):
      return 10

# 자식클래스
class Child(Parent): # Child는 Parent를 상속 받는다. 상속관계 = is ~ a = Child는 Parent이다
  x = 'child x'

  def display(self):
    print('display')
    #print(Parent.test(self))
    print(Parent.output())       # 상속을 받ㅇ
    print(Child.output())
`````

`````python
print(Parent.x)
print(Parent.y)
`````

`````python
print(Child.x)     # Child에 x값이 있어서 상속값  x
print(Child.y)     # Child에 y값이 없음으로 Parent에서 상속받은 값
`````

`````python
# 아래랑 같은 의미
c = Child()
c.display()
`````

`````python
Child.display(c)
`````

`````python
c.output()
`````

`````python
c.test()
`````

`````python
class Base :
  def __init__(self):
    self.x = 'Base' #추가된부분
    self.y = 'Base' #추가된부분
    print('Base')

class A(Base):
  a = 1

  def __init__(self):
    Base.__init__(self)
    self.x = 'A'          #추가
    print("A")

class B(Base):
  b = 1

  def __init__(self):
    Base.__init__(self)
    self.y = 'B'          #추가
    print("B")

class C(A,B):
  def __init__(self):
    A.__init__(self)
    B.__init__(self)
    print('C')
`````

`````python
c=C()
`````

Base의 생성자가 한번 더 실행되었기 때문에 A를 상속 받았음에도 x가 A가 아니다

`````python
c.x
`````

`````python
c.y
`````

super을 사용하면 다중상속 문제를 해결할 수 있다

`````python
class Base :
  def __init__(self):
    self.x = 'Base' #추가된부분
    self.y = 'Base' #추가된부분
    print('Base')

class A(Base):
  a = 1

  def __init__(self):
    Base.__init__(self)
    self.x = 'A'          #추가
    print("A")

class B(Base):
  b = 1

  def __init__(self):
    Base.__init__(self)
    self.y = 'B'          #추가
    print("B")

class C(A,B):
  def __init__(self):
    #A.__init__(self)
    #B.__init__(self)
    super(C,self).__init__() #suoer사용 super은 부모를 의미한다
    print('C')
`````

`````python
c=C()
`````

Base의 생성자는 한 번만 실행되었다. 그러나 A생성자만 실행되고 B생성자는 실행되지 않았다

`````python
c.x
`````

A만 2번 실행 되었고 Base도 2번 중복 실행 되었다.

`````python
class Base :
  def __init__(self):
    self.x = 'Base' #추가된부분
    self.y = 'Base' #추가된부분
    print('Base')

class A(Base):
  a = 1

  def __init__(self):
    Base.__init__(self)
    self.x = 'A'          #추가
    print("A")

class B(Base):
  b = 1

  def __init__(self):
    Base.__init__(self)
    self.y = 'B'          #추가
    print("B")

class C(A,B):
  def __init__(self):
    #A.__init__(self)
    #B.__init__(self)
    super(C,self).__init__() #suoer사용 super은 부모를 의미한다
    super(C,self).__init__()
    print('C')
`````

`````python
c=C()
`````

`````python
class Base :
  def __init__(self):
    self.x = 'Base' #추가된부분
    self.y = 'Base' #추가된부분
    print('Base')

class A(Base):
  a = 1

  def __init__(self):
    #Base.__init__(self)
    super().__init__()
    self.x = 'A'          #추가
    print("A")

class B(Base):
  b = 1

  def __init__(self):
    #Base.__init__(self)
    super().__init__()
    self.y = 'B'          #추가
    print("B")

class C(A,B):
  def __init__(self):
    #A.__init__(self)
    #B.__init__(self)

    super().__init__()      #python3에서는 생략이 가능
    print('C')
`````
