## 그루핑
 문자열이 계속해서 반복되는지 조사하는 정규식

( ): 그룹을 만들어 주는 메타문자이다.   

group()   
group(0)   
group(1)   
group(2)   

```python
import re
```

```python
p = re.compile(r'(ABC)+')
m = p.search('ABCABCABC OK!')
print(m)
print(m.group(), m.start(), m.end())
print(m.group(0), m.start(), m.end())       # 정규식 전체값 리턴
print(m.group(1) , m.start(), m.end())      # 1부턴 그룹지어준거 리턴
#print(m.group(2))       # 인덱스 에러 발생
```

```python
r = re.compile('\w+\s+(\d+[-]?){3}')      # 모든 문자 공백 모든숫자     ? - 한개 또는 0개
m = r.search('park 010-1234-5678')
print(m)
print(m.group(), m.start(), m.end())
print(m.group(0), m.start(), m.end())       # 정규식 전체값 리턴
print(m.group(1) , m.start(), m.end())
```

```python
r = re.compile('(\w+)\s+((\d)+-\d+-\d+)')
m = r.search('park 010-1234-5678')
print(m)
print(m.group(), m.start(), m.end())
print(m.group(0), m.start(), m.end())       # 그루핑 순서는 왼쪽에서 오른쪽 밖에서 안쪽
print(m.group(1) , m.start(), m.end())
print(m.group(2), m.start(), m.end())
print(m.group(3), m.start(), m.end())
```

```python
!?P<그룹명>
r = re.compile('(?P<name>\w+)\s+(?P<phone>(?P<ko>\d)+-\d+-\d+)')
m = r.search('park 010-1234-5678')
print(m)
print(m.group(), m.start(), m.end())
print(m.group(0), m.start(), m.end())       # 그루핑 순서는 왼쪽에서 오른쪽 밖에서 안쪽
print(m.group(1) , m.start(), m.end())
print(m.group(2), m.start(), m.end())
print(m.group(3), m.start(), m.end())
print(m.group('name'))
```

```python
r = re.compile('.+:')
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('.')
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('.:')
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('.+(?=:)')               # 물음표만 가지고 오고 싶다
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('(?=:).+')               # 물음표만 가지고 오고 싶다
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('(?=.*:)')               # : 가지고 있으면 리턴 x
m = r.search('http://google.com')
print(m)
```

```python
r = re.compile('(?=.*:).+://')               #  () 소멸되지 않는다 앞에 이런 문자가 있는 결과를 찾는다 이거 뒤에 해당하는값을 가져온다
m = r.search('http://google.com')            # 검색을하려면 .*가 들어가야 한다
print(m)
```

```python
r = re.compile('[:].+')               #  [] 는 소멸문자
m = r.search('http://google.com')
print(m)
```
