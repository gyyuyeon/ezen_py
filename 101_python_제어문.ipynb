````python
#while 문을 강제적으로 빠져나왔기 때문에 else 영역을 실행하지 않는다
#break는 반복문을 강제적으로 빠져나올때 사용한다 즉 반복문을 중단 한다
a = 10
while a>5:
  a-= 1 #a= a-1
  print(a)
else :
  print('end')
````

````python
a = 10
while a>5:
  a-= 1 #a= a-1
  print(a)
  if a==5:
    break
else :
  print('end')
````

````python
# continue 는 다음 차례 루프로 넘어간다
# 반복문을 중다나지 않음으로 else문이 실행된다
a = 10
while a>5 :
  a = a -1
  if a == 6:
    continue  # 반복문에서 continue를 만나면 무조건 조건식으로 간다 밑에꺼 재생x
  print(a)
else :
  print('end')
````

### enumerate 함수
 - enumerate()함수는 인덱스와 객체를 동시에 저장할 수 있는 특징을 가진 for문을 좀더 효율적으로 사용하기 위한 함수이다.

````python
for k in enumerate(['appke','python','microsoft']):
  print(k)
````

````python
for k,v in enumerate(['appke','python','microsoft']):
  print(k,v)
````

## List Compression
- List 정의시 값을 정하지 않고 호출시 리스트 내의 값들이 처리되도록 구성
- A = [표현식 for 항목 in 반복 가능 객체(iterable)]
- A = [표현식 for 항목 in 반복 가능 객체(iterable) if 논리식]
- A = [표현식 for 항목 in 반복 가능 객체(iterable)1 for 항목2 in 반복 가능 객체(iterable)2 if 논리식 ]

````python
#제곱
s = []
for x in range(10) :
  s.append(x**2)
print(s)
````

````python
s = [x**2 for x in range(10)]
print(s)
````

````python
s = [x**2 for x in range(10) if x% 2==0]
print(s)
````

````python
s =[]
for i in range(1,3):
  for j in range(4,8):
    print('{}{}'.format(i,j))
    s.append((i,j))
print(s)
````

````python
s =[(i,j) for i in range(1,3) for j in range(4,8)]
print(s)
````

````python

s = []
cnt = [1,1,2,1]
prod = [500,1000,400,600]
for i,j in zip(cnt,prod):
  s.append(i*j)

print(sum(s))
````

````python

cnt = [1,1,2,1]
prod = [500,1000,400,600]
s = sum([i*j for i,j in zip(cnt,prod)])
print(s)
````

### dict comprehension
- 사전 정의시 값을 정하지 않고 호출시 사전 내의 값들이 처리되도록 구성
- A = {표현식for(k,v) in 반복 가능 객체 if 논리식}

````python
d = {}
for k,v in zip(range(9),range(9)) :
  d[k]=v
print(d)
````

````python
d = { k:v for k,v in zip(range(9),range(9))}
d
````

````python
a = {'x':1,'y':2,'z':3}
b = {'w':10,'x':11,'y':2}
print(a.keys())
print(b.keys())
c = a.keys() - b.keys()
print(c)
d = c.pop()
print('{}:{}'.format(d,a[d]))  # pop은 완전히 꺼내오는거라 2번 불가능 # 변수에 따로 c.pop()
````

````python
a = {'x':1,'y':2,'z':3}
b = {'w':10,'x':11,'y':2}
k = {key:a[key] for key in a.keys() - b.keys()}
print(k)
````

````python
k ={ (i,j) for i in range(1,3) for j in range(4,8)} #튜플로 묶어서 call
print(k)
````
