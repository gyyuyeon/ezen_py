```python
class InvalidException(Exception):


  def __init(self):
    super().__init__('입력값에 오류가 있습니다.')


```

```python
class Calculator :


  @staticmethod           # 외부에서 class.method로 호출할수있다 객체를 self 로 지정하지 않는다
  def getSum(data):       # 데이터 입력 값 개수를 늘리려면 여기다가 #  def getSum(data1, data2,data3):
      total = 0.0


      try:
        if(data <2 or data>5):
          raise InvalidException
        else :
          for i in range(1,data+1):
            total += i

      except InvalidException as e:
        print(e)
        print('2보다 크거나 같고 5보다 작거나 같아야 한다')

      return total

if __name__ == '__main__':
  data = int(input('숫자입력:'))
  res = Calculator.getSum(data)
 #data = input('이름')
```
