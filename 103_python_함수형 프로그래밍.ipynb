## 순수함수(pure function) & 부수효과(side effect)
 - 순수함수 : 동일한 입력 값에 항상 동일한 출력값을 나타냄
 - 부수효과 : 리턴값으로 결과를 만드는 것 외에 들어온 값을 직접 변경하는 등 외부에 영향을 주는것

```python
# 비순수함수
# 매개변수 값을 변경함으로 순수함수가 아니다
def func(a) :
  a.append(3)
  a.append(4)
  return a
```

```python
func(1,2)
```

```python
c = 14
func(1,2)
```

```python
def func(a,b) :
  return a+b
```

## 고차함수(higher order function)
 - 함수를 값으로 사용한다
 - 함수를 parameter에서 받을수 있다
 - 함수를 리턴할수 있다
 - map, filter, reduce는 함수형 프로그래밍에서 자주 사용되는 고차원 함수이다. 이중에서 reduce는 functool모듈을 import 해야만 사용할 수 있도록 바뀌었다.

 ### 일급객체(first class object)
 - 변수에 할당 가능
 - 다른 함수의 전달인자로 전달 가능해야하고
 - 다른 함수의 결과로서 리턴 가능해야한다

  *** 함수는 일급객체이므로 고차함수를 사용할 수 있다 ***

```python
def ss():
  return 3
```

```python
def display(a):
  return a
```

```python
# ss() 넘겨주면 display는 고차함수가 아니다.
display(ss())
```

```python
#display는 고차함수
k = display(ss)
k()
```

```python
display(ss)() #앞에꺼부터 처리하고 뒤에꺼 처리
```

##callback(콜백)
- 함수 파라미터에서 다른 함수를 받으면 이 파라미터를 콜백함수라고 이야기한다

```python
def plus(callback):
  callback()
```

```python
# 애가 callback함수  # 이벤트가 발생했을떄
def display() :
  print('display')
```

```python
plus(display)
```

## closur(클로저)
 - 함수의 중첩 : 함수안에 함수를 정의할 수 있다.
 - 함수를 중첩해서 사용하는 기법이다.
 - 함수를 정의할 때 outer함수는 inner함수를 return하고, inner함수에서는 outer함수의 자원을 사용한다.

```python
#내부에서 outer함수에있는 자원을 사용하기 위한 목적으로 사용
def t(a):
  def x():
    return a+1
  return x
```

```python
p = t(3)
p()
```

```python
def outer():
  isShow = False
  def inner():
    nonlocal isShow
    isShow = not isShow
    return isShow
  return inner
```

```python
k = outer()
k()
```

```python
# 클로저
def outer(isShow):
  def inner():
    return not isShow   # outer값을 참조한걸 넘김
  return inner
```

```python
k = outer(False)
k()
```

```python
# outer와 inner 값 둘다 참조
def outer(m):
  def inner(n):
    return m+n
  return inner
```

```python
outer(3)(4)
```
