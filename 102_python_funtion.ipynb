특정년도의 윤년,평년을 구해서 윤년이면 True 평년이면 False 리턴하는 프로그램을 구현하세요.

````python
def isLeapYear(year):
  if (year %4 ==0 and year%100 != 0) or year%400 == 0 :
    return True
  else :
    return False
````

외부함수의 자원을 내부함수에서 참조할때는 nonlocal을 사용한다

````python
k = 1

def outer():
  m=1
  def inner():
    global k
    k=k+1
    nonlocal m
    m=m+1
    return k,m
  return inner()
````
