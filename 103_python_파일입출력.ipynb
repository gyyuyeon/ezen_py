## 파일 입출력
open('파일경로','파일열기모드')
- w : 쓰기모드 - 파일에 내용을 쓸때 사용
- r : 읽기모드 - 파일에 내용을 읽을때 사용
- a : 추가모드 - 파일에 내용을 추가할때 사용

````python
fwrite = open('source/sample.txt','w')#파일 열기
for i in range(5):
  data = '{}번째 줄입니다.\n'.format(i)
  fwrite.write(data)
fwrite.close()   #모든 입출력에 해당하는거 할때는 끝나고 닫아야 한다
````

````python
#첫줄만 읽기
fread = open('source/sample.txt','r')
line = fread.readline() # 한라인만읽는다
print(line)
print(type(line))
fread.close()
````

````python
#한줄씩 반복
fread = open('source/sample.txt','r')  #파일을 읽기로 연결한다
while True :
  line = fread.readline() # readline은 한라인만읽는다 #파일의 끝이면 empty값을 리턴해줌
  if not line :           # 파일에서 더이상 데이터 값이 없을때(읽어올 내용이 없을때) 조건 만족
    break
  print(line,end='')      # 파일 끝에 엔터
fread.close()
````

````python
# readlines 한줄씩 읽어와서 리스트에 저장한후 리스트 값으로 출력
fread = open('source/sample.txt','r')  #파일을 읽기로 연결한다
while True :
  line = fread.readlines() # readline은 한라인만읽는다 #파일의 끝이면 empty값을 리턴해줌
  if not line :           # 파일에서 더이상 데이터 값이 없을때(읽어올 내용이 없을때) 조건 만족
    break
  print(line,end='')      # 파일 끝에 엔터
  print(type(line))
fread.close()
````

````python
# 기존파일에 새로운 데이터를 추가할 때에는 모드의 값을 'a'로 설정한다
fwrite = open('source/sample.txt','a')#파일 열기   #append 추가 덮어쓰기
for i in range(11,21):
  data = '{}번째 줄입니다.\n'.format(i)
  fwrite.write(data)
fwrite.close()   #모든 입출력에 해당하는거 할때는 끝나고 닫아야 한다
````

## with문과 함께 사용
 - with문과 함께 파일을 사용하면 읽기, 쓰기할 때 close()를 안해도된다.


````python
with open('source/sample.txt','w') as f:
  for i in range(1, 11):
    data = '{}번째 줄입니다.\n'.format(i)
    f.write(data)
````

````python
with open('source/sample.txt','r') as f:
  while True:
    line = f.readline()
    if not line:
      break
    print(line)
````

jumsu.txt파일의 내용을 읽어와서 오름차순으로 정렬을 한후 출력하는 프로그램을 구현하세요

````python
with open('source/jumsu.txt','r') as f:
  line = f.read()
data = line.split(',')

hap =0
for i,v in enumerate(data) :
  data[i] = int(v)
  hap += data[i]

print(sorted(data))         # 내림차순 k.sort() = 리턴 없음 a = sorted(k) a=리턴값
print('합계는:',hap)
print('평균은:',hap/len(data))
````

````python
# 문1-2
with open('source/score.txt','r') as f :
  for line in f:
    total = 0
    name, kor, mat, eng = line.strip().split('/')   # 언팩킹 작업
    print(line.strip().split('/'))

# 리스트로 받는거 확인
# ['kim', '56', '78', '12']
# ['hong', '46', '100', '97']
# ['park', '96', '56', '88']
````

````python
#문 1-3
with open('source/score.txt','r') as f :
  for line in f:
    total = 0
    name, kor, mat, eng = line.strip().split('/')   # 언팩킹 작업
    # 언팩킹해서 받아옴 =  line을.공백제거.슬레시로 구분(리스트형태)
    total = int(kor)+int(mat)+int(eng)              # total = sum(int(kor).int(mat).int(eng))
    print('{:>6}의 점수 합계 : {:<5}'.format(name,total))  # {:5}5자리

#kim의 점수 합계 : 146
# hong의 점수 합계 : 243
# park의 점수 합계 : 240
````

[문제3]
input.txt 파일에는 팝송 가사가 들어있다. 이 파일에서 검색하고자 하는 문자열이 포함되어있는 라인의 번호와 가사를 출력하는 프로그램을 구현하세요. 만약에 'you'가 포함된 문장을 출력한다면 대소문자 구분없이 찾도록 한다.

[출력결과]
5 line : It exists to give You comfort
6 line : It is there to keep you warm
9 line : When You are most alone
10 line : The memory of love will bring you home
14 line : It invites you to come closer
15 line : It wants to show you more
17 line : And even if you lose yourself
20 line : will see you through
39 line : My memories of love will be of you

````python
#문3-1 멤버인체크 사용
with open('source/input.txt','r') as f :
  search = input('search:').lower()      # input하기
  for i,v in enumerate(f):
    if v.lower().find(search) != -1 :
      print('{} line : {}'.format(i+1,v),end='')
````
