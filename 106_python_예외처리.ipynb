## 예외처리
- 예외처리는 에러가 발생하되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다
- 예외처리구조
```
   try:
      실행할 코드
   except:
      오류가 발생했을때 처리하는 코드
   else:
      오류가 없을때 처리하는 코드
   finally:
      무조건 처리하는 코드

```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할때 에러가 발생하면 디버깅 하는 스타일 이다.
- EAFP
 - 허락보다는 용서를 구하기 쉽다.(Easier to ask for forgovenees than permission)

- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을때만 except를 발생 시킨다
- Built-in Ecepttions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy

````python
# try 블록이 정상적으로 실행디지 않으면 else 블록은 실행되지 않는다.
# try 실행 -> except 실행 -> finally 실행

try :
  a = int(3)
  b = int('a')
  c = int(4)
except:
  print('숫자가 아닌 문자가 입력되어 있음')

else:
  print('{}와{}와{}의 합:{}'.format(a,b,c,a+b+c)))

finally:
  print('program end')
````

````python
# try 블록이 완벽하게 실행이 되었기 때문에 else 블록이 실행이 된다.
# try 실행 -> else 실행 -> finally 실행
try :
  a = int(3)
  b = int(5)
  c = int(4)
except:
  print('숫자가 아닌 문자가 입력되어 있음')

else:
  print('{}와{}와{}의 합:{}'.format(a,b,c,a+b+c))

finally:
  print('program end')
````

## 다중 except
- 다중 except 선언이 가능하다.
- except 뒤에 오는 as 는 try 블록에서 발생한 이름을 받아온다.

```
try:
      실행할 코드
   except 예외 as 변수:
      예외가 발생했을때 처리하는 코드
   except 예외 as 변수:
      예외가 발생했을때 처리하는 코드
   except 예외 as 변수:
      예외가 발생했을때 처리하는 코드
   


```

````python
y = [10,20,30]

try:
  index , x =map(int,input('인덱스와 나눌 숫자를 입력하세요:').split())     # string 으로 받는걸 split으로 받으면  list로 리턴됨
  print(y[index]/x)
  #index , x =map(int,['2','3'])

except ZeroDivisionError as e:
  print('숫자를 0으로 나눌수 없습니다',e)
except IndexError as e:
  print('잘못된 인덱스 입니다',e)
except Exception as e:
  print(e)
else:
  print('end')
finally:
  print('program end')
````

## 사용자 정의 예외 클래스 만들기

````python
class UserClass(Exception):
  def __str__(self):
    return '허용되지 않은 데이터 입니다.'
````

````python
try:
  x = int(input('숫자입력:'))
  if x<1:
    raise UserClass
  print(x)
except UserClass as e:
  print(e)
````
