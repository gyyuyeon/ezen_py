```python
a = [[1,2,3],[4,5,6]] # python에서 제공하는 데이터 타입인 list
b = a

print(a is b)      # 같은 주소
print(a == b)

a[0][0] = 10
print(a is b)
print(a == b)
```

```python
a = [[1,2,3],[4,5,6]] # python에서 제공하는 데이터 타입인 list
c = a.copy()
print(a is c)
print(a == c)

a[0][0] = 10
print(a is c)
print(a == c)
```

```python
import copy
a = [[1,2,3],[4,5,6]] # python에서 제공하는 데이터 타입인 list
d = copy.deepcopy(a)
print(a is d)
print(a == d)

a[0][0] = 10
print(d[0] is a[0])
print(d[0] == a[0])
```
