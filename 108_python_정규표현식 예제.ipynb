문제1] 이름과 년도을 출력하도록 구현

['(이동민, 2019)', '(최재영, 2019)', '(Lion, 2018)']

```python
example1 = '이동민 교수님은 다음과 같이 설명했습니다(이동민, 2019). 그런데 다른 교수님은 이 문제에 대해서 다른 견해를 가지고 있었습니다(최재영, 2019). 또 다른 견해도 있었습니다(Lion, 2018)'
```

```python
#re.findall('정규표현식',대상문자)
re.findall('\([가-힣a-zA-Z]+,\s+\d{4}\)',example1)   # 괄호취급 한글/영어소문자/영어대문자(한개이상),공백/숫자(4글자) 괄호, 이안에서
```

```python
re.findall('([가-힣a-zA-Z]+,\s+\d{4})',example1)
```

문제12] words리스트에 저장된 여러 단어중 'a'로 시작하는 단어만 출력  

apple  
arise  
above  

```python
words = ['apple', 'cat', 'brave', 'drama', 'arise', 'blow', 'coat', 'above']
```

```python
for i in words:
  m = re.match(r'a\D+',i)  # \d  = 숫자   \D  = 숫자가 아닌   + 한개이상
  if m :
    print(m.group())
```

문제14] 숫자에서  천단위를 제거


['123강남', 151767.0, 11093.0, 27394.0]

```python
data = ['123강남', '151,767', '11,093', '27,394']
```

```python
k = []

for j in data:
  if re.search('[a-z가-힣]',j):
    k.append(j)
  else:
    k.append(float(re.sub(',','',j)))

k
```

문제14] 이메일 형식 체크  
  <조건>   
  1 min@daum.net  
  2 min@seoul.go.kr

```python
email = input('이메일:')
r = re.compile('^[a-zA-Z][\w]*@[a-z]+.[a-zA-Z0-9]+[.[a-zA-Z]{2,3}]?$')
m = re.search(r, email)
print(m)
```

문제15] 아이디 정규 표현식<조건>

1. 전체길이 5~8자
2. 영문자로 시작
3. 숫자가 2개이상포함 되어있어야함
4. 2자 이상의 반복 문자가 연속 2번있으면 안됨

```python
id = 'kwkd24i'
r = re.compile(r'^(?=.*\d+.*\d+)(?!.*(\w{2})\1)[a-zA-Z][a-zA-Z0-9]{4,7}$')   #중간에 저건 첫영문자 뒤에 숫자가 있어야한다(2개이상)
                                                                 #?= 긍정 전방탐색
                                                                 #?! 부정전방 탐색 인덱스 1이 반복 되면 안됨
m = re.search(r,id)
print(m)
```
