````python
class Sales:
    def __init__(self, item=None, qty=None, cost=None):
      self.__item = item
      self.__qty = qty
      self.__cost = cost

    #getter

    def getItem(self):
      return self.__item

    def getQty(self):
      return self.__qty

    def getCost(self):
      return self.__cost


    #setter
    def setItem(self, item):
      self.__item = item

    def setQty(self ,qty):
      self.__qty = qty

    def setCost(self, cost):
      self.__cost = cost



    def getPrice(self):                  # 멤버변수
      return self.__qty * self.__cost    # 둘다 int 값이라 그냥 곱하면 된다

    def toString(self):
      return '{}{}원 짜리 {}개 구입하면 {}원이 필요함'.format(self.__item, self.__qty, self.__cost, self.getPrice())
      # 메소드는 괄호및 언더바 필요x
      # 안되면 부분적으로 찾으면서 들여쓰기
      # 못찾는다 오타 및 들여쓰기





if __name__ == '__main__':

  ss = Sales()
  ss.setItem('Apple')
  ss.setQty(5)
  ss.setCost(1200)
  print(ss.toString())

````

```
 문제)
 사람의 신장(H)와 몸무게(W)를 입력받아 비만도(B)를 계산하는 프로그램작성
 표준체중:SW=(H-100)*0.9
 비만도: B(%)=(W-SW)/SW*100

 HealthRate 클래스에 아래 사항을 추가합니다.
 1 표준체중 계산하는 메소드를 정의한다. standardHealth(self)
 2 비만도를 계산하는 메소드를 정의한다.  rateCheck(self)
 비만도가 10% 이내이면 정상, 10%이상 20%이면 과체중, 20% 이상은 비만


 3 출력결과
 홍길동님의 신장 168 , 몸무게 45  정상입니다.
 박둘이님의 신장 168 , 몸무게 90  비만입니다.
```

````python
# 다른 클래스에서 생성자만 받아옴
class Health:

  def __init__(self, name, height, weight):
    self.name = name;
    self.height = height;
    self.weight = weight;


class HealthRate:


  name = input('이름:')
  height = int(input('신장:'))
  weight = int(input('몸무게:'))
  def standardHealth(self):
    return (self.height-100)*0.9

  def rateCheck(self):
    B=(self.weight-self.standardHealth())/self.standardHealth()*100
    if B < 10:
      return '정상'
    elif B>=10 and B<20:
      return '과체중'
    else :
      return '비만'


  def toString(self):
    return '{}님의 신장 {}, 몸무게{} {}입니다'.format(self.name, self.height, self.weight, self.rateCheck())


if __name__ == '__main__':
  hr = HealthRate()
  print(hr.toString())
````
